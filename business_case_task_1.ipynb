```python
import pandas as pd
import numpy as np
```

```python
#Load and merge all files Table_* using pandas package
file1 = "C:\\Users\\LocalSystem\\Downloads\\Python_test\\Python_test\\Table_0.xlsx"
file2 ="C:\\Users\\LocalSystem\\Downloads\\Python_test\\Python_test\\Table_1.csv"
file3 = "C:\\Users\\LocalSystem\\Downloads\\Python_test\\Python_test\\Table_2.xlsx"
```

```python
df = pd.concat([pd.read_excel(file1), pd.read_csv(file2), pd.read_excel(file3)], ignore_index=True)
print(df)
```

```python
#Drop duplicates if any appear in data
source_df = pd.DataFrame(df)
result_df = source_df.drop_duplicates()
print('Result DataFrame:\n', result_df)
```

```python
# Show (for example print) number of null values in each column
blank_counts = df.isnull().sum()
print("Number of blank spaces in each column:",blank_counts)
```

```python
#Fill numerical null values with 1337
df.fillna(1337, inplace=True)
df.loc[295:305]
```

```python
#For each row calculate difference (number of days) between Acctg Date and Date columns
df['Acctg Date'] = pd.to_datetime(df['Acctg Date'])
df['Date'] = pd.to_datetime(df['Date'])

df['Date_Difference'] = (df['Acctg Date'] - df['Date']).dt.days
print(df)
```

```python
#Convert Amount to PLN using FXrates.csv
fxrates = pd.read_csv("C:\\Users\\LocalSystem\\Downloads\\Python_test\\Python_test\\FXrates.csv")
df = pd.merge(df, fxrates, on='Currency', suffixes=('_df', '_fxrates'))
df['Amount PLN'] = df['Amount'] * fxrates['Per USD']
```

```python
# Create folder results and save separate file for EACH unique Type value, use Type in name of saved file (for example Table_4P4.xlsx)
i =0;
for type_value in df['Type'].unique():
    
    #if(type_value != '?45'):
    df['Type'] = df['Type'].str.replace('?','')
    all_merged_df = df[df['Type'] == type_value]
    all_merged_df.to_excel(f'Table_{type_value}.xlsx', index=False)
    print(f'Table_{type_value}.xlsx')
```
